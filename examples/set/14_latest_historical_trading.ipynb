# Latest Historical Trading

Fetch the **latest completed trading-day summary** for any SET stock — OHLC prices,
volume/value, and valuation metrics (P/E, P/BV, dividend yield, market cap, book value).

Endpoint: `GET /api/set/stock/{symbol}/latest-historical-trading`

## Setup

```python
# !pip install settfex

from settfex.services.set import Stock, get_latest_historical_trading
```

## 1. Fetch the latest trading day

`get_latest_historical_trading` returns a single `LatestHistoricalTrading` record for the
most recent completed session.

```python
d = await get_latest_historical_trading("CPALL")

print(f"Date  : {d.date:%Y-%m-%d}")
print(f"OHLC  : O={d.open} H={d.high} L={d.low} C={d.close}")
print(f"Change: {d.change} ({d.percent_change:+.2f}%)")
print(f"Volume: {d.total_volume:,.0f} shares")
print(f"Value : {d.total_value:,.0f} THB")
```

## 2. Valuation metrics

The same record carries valuation fields. Note the ratio fields are **`pe`** and **`pbv`**
(not `pe_ratio` / `pb_ratio`).

```python
print(f"P/E              : {d.pe}")
print(f"P/BV             : {d.pbv}")
print(f"Dividend yield   : {d.dividend_yield}%")
print(f"Market cap       : {d.market_cap:,.0f} THB")
print(f"Book value/share : {d.book_value_per_share} THB")
print(f"Listed shares    : {d.listed_share:,.0f}")
```

## 3. Compare several symbols

With the optional `dataframe` extra (`pip install "settfex[dataframe]"`) you can build a
quick comparison table.

```python
symbols = ["CPALL", "PTT", "KBANK", "AOT"]

rows = []
for sym in symbols:
    d = await get_latest_historical_trading(sym)
    rows.append(
        {
            "symbol": d.symbol,
            "close": d.close,
            "pct_change": d.percent_change,
            "pe": d.pe,
            "pbv": d.pbv,
            "div_yield": d.dividend_yield,
            "market_cap": d.market_cap,
        }
    )

try:
    import pandas as pd

    display(pd.DataFrame(rows).set_index("symbol"))
except ImportError:
    for r in rows:
        print(r)
```

## 4. Using the unified `Stock` class

```python
stock = Stock("PTT")

d = await stock.get_latest_historical_trading()

print(f"{d.symbol}: close {d.close} ({d.percent_change:+.2f}%), P/E {d.pe}, P/BV {d.pbv}")
```

## Summary

- `get_latest_historical_trading(symbol)` → one `LatestHistoricalTrading` record.
- OHLCV + change, plus valuation: `pe`, `pbv`, `dividend_yield`, `market_cap`,
  `book_value_per_share`, `listed_share`.
- Also on the unified `Stock` class via `stock.get_latest_historical_trading()`.

See `docs/settfex/services/set/latest_historical_trading.md` for the full reference.
